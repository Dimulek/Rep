```python
groupOfRows = df.groupby(["Name", "Platform", "Publisher"])["Year"].value_counts() #Групировка строк датасета по столбцам 
#Производитель и Год
groupOfRows.where(groupOfRows > 1).dropna()
#Вывод записей имеющие дубликаты дубликатов
```

```python
#Обработка датасета
nonDublDF = df.drop_duplicates(subset=["Name", "Platform", "Publisher"])
nonDublDF #Было 553 Стало 546
```

```python
groupOfRows = nonDublDF.groupby(["Name", "Platform", "Publisher"])["Year"].value_counts()
#Групировка строк датасета по столбцам Производитель и Год
groupOfRows.where(groupOfRows > 1).dropna()
#Вывод записей имеющие дубликаты дубликатов (Их нет)
```

```python
from sklearn.impute import SimpleImputer

simp = SimpleImputer(missing_values=np.nan)
simp = simp.fit(df[['Year']])
df['Year'] = simp.transform(df[['Year']])
df = df.replace({'Publisher' : np.nan}, 'Nothing')
df.isnull().mean()*100
```
