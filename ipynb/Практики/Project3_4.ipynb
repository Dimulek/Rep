```python
from sklearn.preprocessing import LabelEncoder
Lenc = LabelEncoder()
Lenc.fit(df['Genre'])
#df['Genre'] = Lenc.fit_transform(df['Genre'])
Lenc.classes_
```

```python
from sklearn.preprocessing import OneHotEncoder
OnHot = OneHotEncoder()
OnHot.fit(df[['Genre']])
df
```

```python
OnHot.categories_
```

```python
OnHot.transform(df[['Genre']]).toarray()
```

```python
df[['IsAction', 'IsAdventure', 'IsFighting', 'IsMisc', 'IsPlatform', 'IsPuzzle', 'IsRacing', 'IsRole-Playing', 'IsShooter', 'IsSimulation', 'IsSports', 'IsStrategy']] = OnHot.fit_transform(df[['Genre']]).toarray()
#df = df.drop(columns=['IsAction', 'IsAdventure', 'IsFighting', 'IsMisc', 'IsPlatform', 'IsPuzzle', 'IsRacing', 'IsRole-Playing', 'IsShooter', 'IsSimulation', 'IsSports'])
df
```

```python
df = df.drop(columns="Genre")
df
```

```python
from sklearn.preprocessing import Normalizer

Norm = Normalizer()
df[['Global_Sales']] = Norm.fit_transform(df[['Global_Sales']])
```

```python
from sklearn.model_selection import train_test_split
#Other_Sales
X = df[['EU_Sales','Global_Sales']]
Y = df['IsAction']

x_train, x_test, y_train, y_test = train_test_split(X,Y, test_size=0.2,random_state=42)
```

Практика 4

```python
from sklearn.tree import DecisionTreeClassifier

dtc = DecisionTreeClassifier()
```

```python
dtc = dtc.fit(x_train, y_train)
```

```python
dtc_plot = DecisionTreeClassifier(max_depth=40, min_samples_leaf=4, max_leaf_nodes=100)
dtc_plot = dtc_plot.fit(x_train.iloc[::1000], y_train.iloc[::1000])
```

```python
from sklearn import tree
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(35,35))
tree.plot_tree(dtc, filled=True, fontsize=10)
```

```python
from sklearn import metrics
y_predict =dtc.predict(x_test)
```

```python
con_mat = metrics.confusion_matrix(y_test, y_predict)
```

```python
sns.heatmap(con_mat, annot=True, fmt="d")
plt.xlabel("SeeFutur")
plt.ylabel("Real")
```

```python
print("Точность", metrics.accuracy_score(y_test, y_predict))
```

```python
print("Чувствительность", metrics.recall_score(y_test, y_predict))
```

```python
print("Аккуратность", metrics.precision_score(y_test, y_predict))
```

```python
print("Финал", metrics.f1_score(y_test, y_predict))
```
